```python
window = tkinter.Tk()
window.title('Drilling Simulation and Real-Time Optimization Platform')
window.geometry('1000x700')




#========================================
L1 = tkinter.Label(window, 
                   text = '',
                   font=('Arial Bold', 10))
L1.grid(column=1,row=0)
def clicked():
    L1.configure(text='Data Operational parameters Uploaded!', width=50,height=1)

bt1 = tkinter.Button(window, 
                    text="Upload Operational parameters",
                    fg='white',
                    bg='purple',
                    command=clicked, width=30,height=1)
bt1.grid(column=0,row=0)

#========================================
L2 = tkinter.Label(window, 
                   text = '',
                   font=('Arial Bold', 10))
L2.grid(column=1,row=2)

def clicked2():
    L2.configure(text='Data bit details Uploaded!', width=50,height=1)

bt2 = tkinter.Button(window, 
                    text="Upload bit details",
                    fg='white',
                    bg='purple',
                    command=clicked2, width=30,height=1)
bt2.grid(column=0,row=2)

#========================================
L3 = tkinter.Label(window, 
                   text = '',
                   font=('Arial Bold', 10))
L3.grid(column=1,row=3)

def clicked3():
    L3.configure(text='Data Temperature properties Uploaded!', width=50,height=1)

bt3 = tkinter.Button(window, 
                    text="Upload temperature properties",
                    fg='white',
                    bg='purple',
                    command=clicked3, width=30,height=1)
bt3.grid(column=0,row=3)



#========================================
L4 = tkinter.Label(window, 
                   text = '',
                   font=('Arial Bold', 10))
L4.grid(column=1,row=4)

def clicked4():
    L4.configure(text='The Simulator is Running...!', width=50,height=1)

bt4 = tkinter.Button(window, 
                    text="Run the Simulator",
                    fg='white',
                    bg='green',
                    command=clicked4, width=30,height=1)
bt4.grid(column=0,row=4)





#========================================
# L4 = tkinter.Label(window, 
#                    text = '',
#                    font=('Arial Bold', 10))
# L4.grid(column=1,row=4)

# def clicked2():
#     L4.configure(text='Output file Saved!', width=50,height=1)

# bt4 = tkinter.Button(window, 
#                     text="Save output file",
#                     fg='white',
#                     bg='green',
#                     command=clicked2, width=30,height=1)
# bt4.grid(column=0,row=4)























# #========================================

# rad1 = Radiobutton(window, text='PDC ROP Model#1', value=1)
# rad2 = Radiobutton(window, text='PDC ROP Model#2', value=2)
# rad3 = Radiobutton(window, text='PDC ROP Model#3', value=3)
# rad1.grid(column=0,row=4)
# rad2.grid(column=1,row=4)
# rad3.grid(column=2,row=4)

# #========================================

# rad4 = Radiobutton(window, text='Glowka Temprature Model', value=1)
# rad5 = Radiobutton(window, text='New Temprature Model', value=2)
# rad4.grid(column=0,row=5)
# rad5.grid(column=1,row=5)

# #========================================

# rad6 = Radiobutton(window, text='Kerkar et al. Hydraulic Model', value=1)
# rad7 = Radiobutton(window, text='New Hydraulic Model', value=2)
# rad6.grid(column=0,row=6)
# rad7.grid(column=1,row=6)










window.mainloop()
```
